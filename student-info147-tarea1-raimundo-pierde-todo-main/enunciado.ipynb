## Casos confirmados

```python
df = pd.read_csv('./data/covid19_confirmados.csv', index_col = 'Country/Region',parse_dates=True, infer_datetime_format=True)
df.drop(['Lat','Long'],axis='columns', inplace=True)
colname=['Province/State']
for x in df:
    if x!='Province/State':
        x=pd.to_datetime(x)
        colname.append(x)
df.columns=colname
df['Province/State'] = df[pd.to_datetime('2022-04-10')]
df.rename(columns={'Province/State':'Casos totales'}, inplace=True)
df.index.duplicated().sum()
df = df.groupby(df.index).sum()
df
```

## Poblacion por pais

```python
dfp = pd.read_csv('./data/covid19_confirmados.csv', index_col = 'Country/Region',parse_dates=True, infer_datetime_format=True)
dfp.drop(['Lat','Long'],axis='columns', inplace=True)
colname2=['Province/State']
for x in dfp:
    if x!='Province/State':
        x=pd.to_datetime(x)
        colname2.append(x)
dfp.columns=colname2
dfp.index.duplicated().sum()
dfp = dfp.groupby(dfp.index).sum()
dfp
```
